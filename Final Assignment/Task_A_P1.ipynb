### Initializing Scraper

```python
from selenium.webdriver import Chrome 
from instascrape import *
#import "git+https://github.com/stefco/instascrape.git"
```

```python
#change this to your chromedriver location
webdriver = Chrome("./chromedriver")
```

```python
headers = {
    "user-agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Mobile Safari/537.36 Edg/87.0.664.66",
    "cookie": "sessionid=37631852969%3AdMRitrTqpSC4WM%3A8;"
}
imdb = Profile("imdb")
imdb.scrape(headers=headers)
```

### Getting Posts from IMDB

```python
#when you run this cell block, you have to go to the new webdriver page and 
#log into your account in 30 seconds (if you need more time increase the login_pause parameter)
#then you need to stay on the instagram page as it logs you in and loads all the imdb photos

imdb_posts = imdb.get_posts(webdriver=webdriver, amount=200, login_first=True, login_pause=60)
```

```python
len(imdb_posts)
#https://github.com/chris-greening/instascrape/wiki/Scraped-data-points
```

```python
#this takes a long time, avoid rerunning if possible 
#takes 34 minutes to run to completion, do NOT make a smaller pause value, your instagram account will be blocked
scraped_posts_imdb, unscraped_posts_imdb = scrape_posts(imdb_posts, headers=headers, pause=10, silent=False)
```

#### Creating IMDB Dataframe

```python
import pandas as pd

posts_df = pd.DataFrame()
post_id = []
likes = []
captions = []
imageURLs = []
hashtags = []
comments = []
upload_date = []

for p in range(len(scraped_posts_imdb)):
    post_id.append(scraped_posts_imdb[p].shortcode)
    likes.append(scraped_posts_imdb[p].likes)
    captions.append(scraped_posts_imdb[p].caption)
    imageURLs.append(scraped_posts_imdb[p].display_url)
    hashtags.append(scraped_posts_imdb[p].hashtags)
    comments.append(scraped_posts_imdb[p].comments)
    upload_date.append(scraped_posts_imdb[p].upload_date)


posts_df['id'] = post_id
posts_df['likes'] = likes
posts_df['caption'] = captions
posts_df['imageURL'] = imageURLs
posts_df['hashtags'] = hashtags
posts_df['comments'] = comments
posts_df['upload_date'] = upload_date
```

### Getting Posts from Rotten Tomatoes

```python
rottentomatoes = Profile("rottentomatoes")
rottentomatoes.scrape(headers=headers)
```

```python
rt_posts = rottentomatoes.get_posts(webdriver=webdriver, amount=200, login_first=True, login_pause=60)
```

```python
len(rt_posts)
```

```python
scraped_posts_rt, unscraped_posts_rt = scrape_posts(rt_posts, headers=headers, pause=10, silent=False)
```

#### Creating Rotten Tomatoes Dataframe

```python
df_rt = pd.DataFrame()
post_id = []
likes = []
captions = []
imageURLs = []
hashtags = []
comments = []
upload_date = []

for p in range(len(scraped_posts_rt)):
    post_id.append(scraped_posts_rt[p].shortcode)
    likes.append(scraped_posts_rt[p].likes)
    captions.append(scraped_posts_rt[p].caption)
    imageURLs.append(scraped_posts_rt[p].display_url)
    hashtags.append(scraped_posts_rt[p].hashtags)
    comments.append(scraped_posts_rt[p].comments)
    upload_date.append(scraped_posts_rt[p].upload_date)


df_rt['id'] = post_id
df_rt['likes'] = likes
df_rt['caption'] = captions
df_rt['imageURL'] = imageURLs
df_rt['hashtags'] = hashtags
df_rt['comments'] = comments
df_rt['upload_date'] = upload_date
```

```python
df_rt
```

#### Combining IMDB and Rotten Tomatoes Dataframes

```python
#combine the imdb posts and the rotten tomato posts
posts_df.append(df_rt, ignore_index=True)
```

#### Saving Dataframe to CSV file

```python
posts_df.to_csv('posts_df.csv', index=False)
```
