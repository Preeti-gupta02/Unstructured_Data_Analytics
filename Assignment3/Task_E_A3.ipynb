```python
a = posts_with_image_labels['image_labels']
import gensim

def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(token)
    return result

processed_docs = posts_with_image_labels['image_labels'].map(preprocess)
processed_docs[:10]
```

```python
import gensim

dictionary = gensim.corpora.Dictionary(processed_docs)
count = 0
for k, v in dictionary.iteritems():
    print(k, v)
    count += 1
    if count > 10:
        break
```

```python
for k, v in dictionary.iteritems():
    print(k, v)
```

```python
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
bow_corpus[0]
```

```python
processed_docs[0]
bow_doc_0 = bow_corpus[0]

for i in range(len(bow_doc_0)):
    print("Word {} (\"{}\") appears {} time.".format(bow_doc_0[i][0], 
                                               dictionary[bow_doc_0[i][0]], 
bow_doc_0[i][1]))
```

```python
bow_corpus
```

```python
lda_model = gensim.models.LdaMulticore(bow_corpus, num_topics=3, id2word=dictionary, passes=2, workers=2, random_state=100)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
    
topic_1 = 'fashion'
topic_2 = 'photography'
topic_3 = 'human_features'
```

```python
processed_docs[2]
```

```python
for index, score in sorted(lda_model[bow_corpus[2]], key=lambda tup: -1*tup[1]):
    print("\nScore: {}\t \nTopic: {}".format(score, lda_model.print_topic(index, 10)))
```

```python
tab = pd.DataFrame(columns=[topic_1,topic_2,topic_3])
i = 0
for bow_corpus_i in bow_corpus:
    
    for index, score in sorted(lda_model[bow_corpus_i], key=lambda tup: -1*tup[1]):
        #print()
        if index ==0:
        
            tab.loc[i, topic_1] = score
        if index ==1:
        
            tab.loc[i, topic_2] = score
        if index ==2:
        
            tab.loc[i, topic_3] = score
           
    i = i+1
#posts_with_image_labels[fashion]
result = pd.concat([posts_with_image_labels, tab], axis=1)
```

```python
result
```

```python
result_sorted_with_likes = result.sort_values(by=['likes'], ascending=False)
```

```python
result_sorted_with_likes
```

```python
q = result_sorted_with_likes.quantile([0.00, 0.25, 0.50, 0.75, 1.00])
df = result_sorted_with_likes
col = 'likes'

q1 = df[((df[col]>=q[col][0.00]) & (df[col]<q[col][0.25]))]
q4 = df[((df[col]>=q[col][0.75]) & (df[col]<=q[col][1.00]))]

result = pd.concat([q1,q4])

result
#for index, score in sorted(lda_model[bow_corpus[2]], key=lambda tup: -1*tup[1]):
    
```

```python
lower_quartile_avg = q1.mean()
upper_quartile_avg = q4.mean()
```
