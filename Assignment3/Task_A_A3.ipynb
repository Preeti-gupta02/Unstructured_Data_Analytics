## Task A

Scrape Instagram.py to fetch ~200 posts from the Zara Instagram page. This script fetches (i) image URLs, (ii) post caption (the text description of a post), and (iii) # likes.

```python
from selenium.webdriver import Chrome 
from instascrape import Profile, scrape_posts
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
#change this to your chromedriver location
s=Service("./chromedriver")
webdriver = Chrome(service=s)
```

```python
headers = {
    "user-agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Mobile Safari/537.36 Edg/87.0.664.57",
    "cookie": "sessionid=37631852969%3A7LhB3KylDlcAQi%3A2;"
}
zara = Profile("zara")
zara.scrape(headers=headers)
```

```python
#when you run this cell block, you have to go to the new webdriver page and 
#log into your account in 30 seconds (if you need more time increase the login_pause parameter)
#then you need to stay on the instagram page as it logs you in and loads all the zara photos

posts = zara.get_posts(webdriver=webdriver, amount=200, login_first=True, login_pause=60)
```

```python
len(posts)
#https://github.com/chris-greening/instascrape/wiki/Scraped-data-points
```

```python
#this takes a long time, avoid rerunning if possible 
#takes 34 minutes to run to completion, do NOT make a smaller pause value, your instagram account will be blocked
scraped_posts, unscraped_posts = scrape_posts(posts, headers=headers, pause=10, silent=False)
```

```python
import pandas as pd

posts_df = pd.DataFrame()
post_id = []
likes = []
captions = []
imageURLs = []

for p in range(len(scraped_posts)):
    post_id.append(scraped_posts[p].shortcode)
    likes.append(scraped_posts[p].likes)
    captions.append(scraped_posts[p].caption)
    imageURLs.append(scraped_posts[p].display_url)


posts_df['id'] = post_id
posts_df['likes'] = likes
posts_df['caption'] = captions
posts_df['imageURL'] = imageURLs
```
